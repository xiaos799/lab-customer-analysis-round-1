```python
import pandas as pd
```

```python
df1 = pd.read_csv("files_for_lab/csv_files/file1.csv")
df2 = pd.read_csv("files_for_lab/csv_files/file2.csv")
df3 = pd.read_csv("files_for_lab/csv_files/file3.csv")
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)
```

```python
df1 = df1.reindex(columns = ["Customer", "State", "Gender", "Education", "Income", 
 "Vehicle Class", "Policy Type", "Number of Open Complaints", "Monthly Premium Auto", 
 "Total Claim Amount", "Customer Lifetime Value"])
df1.head()
```

```python
df2.head()
```

```python
df2 = df2.rename(columns={"ST":"State", "GENDER":"Gender"})
df2 = df2.reindex(columns = ["Customer", "State", "Gender", "Education", "Income", 
 "Vehicle Class", "Policy Type", "Number of Open Complaints", "Monthly Premium Auto", 
 "Total Claim Amount", "Customer Lifetime Value"])
df2.head()
```

```python
df3.head()
```

```python
df3 = df3.reindex(columns = ["Customer", "State", "Gender", "Education", "Income", 
 "Vehicle Class", "Policy Type", "Number of Open Complaints", "Monthly Premium Auto", 
 "Total Claim Amount", "Customer Lifetime Value"])
df3.head()
```

```python
df = pd.concat([df1,df2,df3])
df.shape
```

numerical columns: Income, Number of Open Complaints, Monthly Premium Auto,	Total Claim Amount,	Customer Lifetime Value

categorical columns: Gender, Education, Vehicle Class, Policy Type

```python
df = df.drop(columns = ["Education", "Number of Open Complaints"])
```

```python
df.head()
```

```python
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].str.rstrip('%')
```

```python
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].astype(float)
```
